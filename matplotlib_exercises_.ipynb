```python
# Create a plot using plt.subplots() and plot X versus X^2 (X squared)
import matplotlib.pyplot as plt
import numpy as np

# 1- إنشاء البيانات
X = np.linspace(0, 100, 100)   # 100 رقم بين 0 و 100
y = X ** 2                     # تربيع كل رقم في X

# 2- إنشاء الشكل والمحور
fig, ax = plt.subplots()

# 3- رسم البيانات
ax.plot(X, y)

# 4- تخصيص الرسم
ax.set_title("X vs X^2")
ax.set_xlabel("X")
ax.set_ylabel("X squared")

# عرض الرسم
plt.show()
```

We'll start with scatter plots.

```python
# Create a scatter plot of X versus the exponential of X (np.exp(X))
import matplotlib.pyplot as plt
import numpy as np

# إنشاء البيانات
X = np.linspace(0, 5, 100)   # نختار 0 لـ 5 عشان np.exp(X) ما يكبرش جدًا
y = np.exp(X)                # e^x

# إنشاء الشكل والمحور
fig, ax = plt.subplots()

# رسم Scatter plot
ax.scatter(X, y, color="red", marker="o")

# تخصيص الشكل
ax.set_title("Scatter Plot of X vs exp(X)")
ax.set_xlabel("X")
ax.set_ylabel("exp(X)")

# عرض الرسم
plt.show()
```

```python
# Create a scatter plot of X versus np.sin(X)
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt

# Create an array of 100 evenly spaced numbers between 0 and 10
X = np.linspace(0, 10, 100)

# Compute the sine of each value in X
Y = np.sin(X)

# Create a scatter plot
plt.scatter(X, Y, color='blue', marker='o', label='sin(X)')

# Add title and labels
plt.title('Scatter Plot of X vs sin(X)')
plt.xlabel('X')
plt.ylabel('sin(X)')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Import the '../data/car-sales.csv' into a DataFame called car_sales and view
import pandas as pd

# Import the CSV file into a DataFrame
car_sales = pd.read_csv('/car-sales.csv')

# View the first 5 rows of the DataFrame
print(car_sales.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Remove non-numeric characters (like '$' or ',') and convert to float
car_sales['Price'] = car_sales['Price'].replace('[\$,]', '', regex=True).astype(float)

# Now plot the 'Price' column
car_sales['Price'].plot()

# Add title and labels
plt.title("Car Prices")
plt.xlabel("Index")
plt.ylabel("Price ($)")
plt.show()

```

Hint: It's not numeric data.

In the process of turning it to numeric data, let's create another column which adds the total amount of sales and another one which shows what date the car was sold.

Hint: To add a column up cumulatively, look up the cumsum() function. And to create a column of dates, look up the date_range() function.

```python
import matplotlib.pyplot as plt

car_sales['Price'].plot()
plt.title("Cleaned Car Prices")
plt.xlabel("Index")
plt.ylabel("Price")
plt.show()

```

```python
# Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column (cumsum())


# Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today)

# View the car_sales DataFrame
import pandas as pd
from datetime import datetime, timedelta

# Example: assume car_sales DataFrame is already loaded and cleaned

# 1. Add 'Total Sales' column using cumulative sum
car_sales['Total Sales'] = car_sales['Price'].cumsum()

# 2. Add 'Sale Date' column with successive dates starting from today
today = datetime.today()
car_sales['Sale Date'] = [today + timedelta(days=i) for i in range(len(car_sales))]

# 3. View the updated DataFrame
print(car_sales.head())
```

Now we've got a numeric column (`Total Sales`) and a dates column (`Sale Date`), let's visualize them.

```python
# Use the plot() function to plot the 'Sale Date' column versus the 'Total Sales' column
import matplotlib.pyplot as plt

# Plot 'Sale Date' vs 'Total Sales'
car_sales.plot(x='Sale Date', y='Total Sales', kind='line', marker='o')

# Add title and labels
plt.title("Cumulative Total Sales Over Time")
plt.xlabel("Sale Date")
plt.ylabel("Total Sales ($)")
plt.xticks(rotation=45)  # Rotate dates for better readability
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Convert the 'Price' column to the integers


# Create a scatter plot of the 'Odometer (KM)' and 'Price' column using the plot() function

import matplotlib.pyplot as plt

# Convert 'Price' to integers
car_sales['Price'] = car_sales['Price'].astype(int)

# Create a scatter plot using the plot() function
car_sales.plot(x='Odometer (KM)', y='Price', kind='scatter', color='blue')

# Add title and labels
plt.title("Price vs Odometer (KM)")
plt.xlabel("Odometer (KM)")
plt.ylabel("Price ($)")
plt.grid(True)

# Show the plot
plt.show()

```
